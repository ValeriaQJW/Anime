```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
df = pd.read_csv('anime.csv')
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['type'], inplace=True)
df.dropna(subset=['genre'], inplace=True)
df.dropna(subset=['rating'], inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
import numpy as np
df = df.sample(5000)
df = df.convert_dtypes(convert_integer=True)
# df['episodes'] = df['episodes'].astype(int)
```

```python
df=df[~df['episodes'].isin(["Unknown"])] 
df['episodes'] = df['episodes'].astype('int')
```

```python
df.to_csv('anime2.csv')
```

```python
df['genre'] = df['genre'].apply(lambda x: str(x).split(','))
df.head(10)
```

```python
df = df.explode('genre')
df
```

```python
df.to_csv('anime3.csv')
```
